```python
import pyedflib
import numpy as np
import pandas as pd
%matplotlib inline
import os
import mne
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import networkx as nx
import sails
```

```python
#read the data
f = pyedflib.EdfReader('S040R01.edf') #read the file
n = f.signals_in_file #number of signals (64)
signal_labels = f.getSignalLabels() #name of the signals
sigbufs = np.zeros((n, f.getNSamples()[0])) #create a empty array
for i in np.arange(n):
    sigbufs[i, :] = f.readSignal(i)  #fill up the arrays with the signal from the edf file
    
#make them into pandas df
eo=pd.DataFrame(sigbufs)
eo.head()
```

```python
#read the data
f = pyedflib.EdfReader('S040R02.edf')
n = f.signals_in_file
signal_labels = f.getSignalLabels()
sigbufs = np.zeros((n, f.getNSamples()[0]))
for i in np.arange(n):
    sigbufs[i, :] = f.readSignal(i)
    
#make them into pandas df
ec=pd.DataFrame(sigbufs)
ec.head()
```

```python
#let's try to use the mne libraru, with lots of tools for eeg
#we need to specify the info of our data to pass to the I/O function in mne libabry

#we have 64 signal of type 'grad' (????) so we create a 1x64 array with this specification
ch_types=['grad']*64
#then we create the info. we pass the signal labels, channel types and the frequency of records per seconds
info = mne.create_info(ch_names=signal_labels, sfreq=1000, ch_types=ch_types)

#now we can make our file into a raw to be plotted
raw=mne.io.RawArray(eo,info)

#scaling (?)
#scalings = { 'grad': 64}

#let's finally plot.
raw.plot(n_channels=64, scalings='auto', title='Data from arrays',
         show=True, block=True)
```

```python
#now let's compute the correlation matrix
eo=eo.transpose()
eo_corr=eo.corr()
#the matrix has several NAN. let's fill them with 0s
eo_corr=eo_corr.fillna(0)
```

```python
#let's see how it looks like 
f = plt.figure(figsize=(7, 7))
plt.matshow(eo_corr, fignum=f.number)
plt.title('Correlation Matrix', fontsize=16);
```

```python
#we need to set a thrashold to create the adjacency matrix from the correlation matrix
#we will maintain only couples which correlation in module is higher than a value k

#create empty matrix to fill
eo_adj=np.zeros((len(eo_corr),len(eo_corr)))

#check whose cells has correlation >k and !=1 (to avoid the diagonal)
for i in range(len(eo_corr)):
    for k in range(i,len(eo_corr)):
        if abs(eo_corr[i][k])>.763 and (eo_corr[i][k])!=1:
            eo_adj[i][k]=1
```

```python
#let's plot that damn adj matrix
f = plt.figure(figsize=(7, 7))
plt.matshow(eo_adj, fignum=f.number)
plt.title('Adjacency Matrix', fontsize=16);
```

```python
#create the graph
eo_g=nx.from_numpy_matrix(eo_adj)

#we need a dictionary with couples 0...64 and signal labels in order to rename the nodes of the graph
mapping={}
for i in range(len(signal_labels)):
    mapping[i]=signal_labels[i]
mapping


#let's rename them
eo_g=nx.relabel_nodes(eo_g, mapping)

#now give them a location 
#open the location file
position=pd.read_csv('channel_locations.txt', sep='\s+')

#create a pos dictionary to pass the position to the nodes of our graph
pos={}
for i in range(64):
    pos[position['label'][i]]=[position['x'][i],position['y'][i]]
```

```python
#the density must be 0.20
nx.density(eo_g)
```

```python
#let's see our graph
f = plt.figure(figsize=(10, 10))
nx.draw(eo_g, pos)
nx.draw_networkx_labels(eo_g, pos=pos, font_color='yellow')
plt.show()
```
